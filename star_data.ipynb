```r
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 10)

star_data <- read_csv("star_classification_small.csv", col_names=TRUE)
colnames(star_data) <- c("right_ascension_angle", "declination_angle", "uv_filter", "green_filter", "red_filter", "near_infared_filter", "infared_filter", "feild_id", "class", "redshift", "plate_id", "date")
star_data

```

```r
plot <- star_data %>%
ggplot(aes(x=right_ascension_angle, y=redshift, color=class))+
    geom_point(size=1, alpha=0.2)
    
plot
```

```r
plot2 <- star_data %>%
ggplot(aes(x=right_ascension_angle, y=declination_angle, color=class))+
    geom_point()
plot2
```

```r
plot3 <- star_data %>%
ggplot(aes(x=redshift, y=declination_angle, color=class))+
    geom_point()
plot3
```
